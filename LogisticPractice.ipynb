```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler,OneHotEncoder
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
df = pd.read_csv('/content/customer_churn_dataset-testing-master.csv')
df.head(5)
```

```python
# setup features and target
x = df.drop(columns=['CustomerID','Churn'])
y = df['Churn']
```

```python
# split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
#encoding
onhot = OneHotEncoder(handle_unknown='ignore',sparse_output=False,drop='first')
x_train_encode = onhot.fit_transform(x_train[['Gender','Subscription Type','Contract Length']])
x_test_encode = onhot.transform(x_test[['Gender','Subscription Type','Contract Length']])
```

```python
#scaling
scaler = StandardScaler()
cols = ['Age','Tenure','Usage Frequency','Support Calls','Payment Delay','Total Spend','Last Interaction']
x_train_scale = scaler.fit_transform(x_train[cols])
x_test_scale = scaler.transform(x_test[cols])
```

```python
x_train_final = np.hstack((x_train_scale,x_train_encode))
x_test_final = np.hstack((x_test_scale,x_test_encode))
```

```python
x_train_final
```

Logistic Regression Algorithm

```python
# train
model = LogisticRegression()
model.fit(x_train_final,y_train)
```

```python
y_pred = model.predict(x_test_final)
```

```python
print('Accuracy:',accuracy_score(y_test,y_pred))
print('Confusion:',confusion_matrix(y_test,y_pred))
print('classificationReport:',classification_report(y_test,y_pred))
```

KNN Algorithm

```python
Kmodel = KNeighborsClassifier(n_neighbors=3)
Kmodel.fit(x_train_final,y_train)
```

```python
y_pred_kNN = Kmodel.predict(x_test_final)
```

```python
print('Accuracy:',accuracy_score(y_test,y_pred_kNN))
print('Confusion:',confusion_matrix(y_test,y_pred_kNN))
print('classificationReport:',classification_report(y_test,y_pred_kNN))
```

```python
# lets try another k numbers too:
for k in [1,2,3,4,5,6]:
  Kmodel = KNeighborsClassifier(n_neighbors=k)
  Kmodel.fit(x_train_final,y_train)
  y_pred_kNN = Kmodel.predict(x_test_final)
  print('KNumber:',k)
  print('Accuracy:',accuracy_score(y_test,y_pred_kNN))
  print('Confusion:',confusion_matrix(y_test,y_pred_kNN))
  print('classificationReport:',classification_report(y_test,y_pred_kNN))
```

SVM(Support Vector Machine)

```python
svcmodel = SVC(kernel='rbf',C=1.0,gamma='scale')
svcmodel.fit(x_train_final,y_train)
```

```python
y_pred_svm = svcmodel.predict(x_test_final)
```

```python
print('Accuracy:',accuracy_score(y_test,y_pred_svm))
print('Confusion:',confusion_matrix(y_test,y_pred_svm))
print('classificationReport:',classification_report(y_test,y_pred_svm))
```

Decision Tree

max_depth = how far split should go

min_sample_split = min data required to split

min_sample_leaf = min data required to produce output otherwise upper split is last

```python
decisionModel = DecisionTreeClassifier(max_depth=5)
decisionModel.fit(x_train_final,y_train)
```

```python
y_pred_decision = decisionModel.predict(x_test_final)
```

```python
print('Accuracy:',accuracy_score(y_test,y_pred_decision))
print('Confusion:',confusion_matrix(y_test,y_pred_decision))
print('classificationReport:',classification_report(y_test,y_pred_decision))
```

```python
print('Accuracy:',accuracy_score(y_train,decisionModel.predict(x_train_final)))
```

Random Forest

```python
RFModel = RandomForestClassifier(n_estimators=100,
                                 max_depth=5,
                                 max_features='sqrt',
                                 min_samples_split=10,
                                 min_samples_leaf=5,
                                 random_state=53)
RFModel.fit(x_train_final,y_train)
```

```python
y_pred_rf = RFModel.predict(x_test_final)
```

```python
print('Accuracy:',accuracy_score(y_test,y_pred_rf))
print('Confusion:',confusion_matrix(y_test,y_pred_rf))
print('classificationReport:',classification_report(y_test,y_pred_rf))
```

```python
print('Accuracy of train:',accuracy_score(y_train,RFModel.predict(x_train_final)))
```
