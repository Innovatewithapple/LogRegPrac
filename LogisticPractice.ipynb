```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
df = pd.read_csv('/content/customer_churn_dataset-testing-master.csv')
```

```python
df.sample(5)
```

```python
x = df.drop(columns='Churn')
y = df['Churn']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
# encoding
onehot = OneHotEncoder(handle_unknown='ignore', drop='first', sparse_output=False)
```

```python
x_train_encode = onehot.fit_transform(x_train[['Gender','Subscription Type','Contract Length']])
x_test_encode = onehot.transform(x_test[['Gender','Subscription Type','Contract Length']])
```

```python
cols = ['CustomerID','Age','Tenure','Usage Frequency','Support Calls','Payment Delay','Total Spend','Last Interaction']
x_train_num = x_train[cols].reset_index(drop=True)
x_test_num = x_test[cols].reset_index(drop=True)

x_train_encode = pd.DataFrame(x_train_encode,columns=onehot.get_feature_names_out())
x_test_encode = pd.DataFrame(x_test_encode,columns=onehot.get_feature_names_out())

x_train_final = pd.concat([x_train_num,x_train_encode],axis=1)
x_test_final = pd.concat([x_test_num,x_test_encode],axis=1)
```

```python
x_train_final
```

```python
# scaling
scale = StandardScaler()
x_train_scale = scale.fit_transform(x_train_final)
x_test_scale = scale.transform(x_test_final)
```

```python
model = LogisticRegression()
model.fit(x_train_scale,y_train)
```

```python
print('coeff:',model.coef_[0])
print('intercept:',model.intercept_)
```

```python
y_pred = model.predict(x_test_scale)
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
